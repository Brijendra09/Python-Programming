# HashTable and HashMap

```python
# Hash tables or has maps in Python are implemented through the built-in dictionary data type. 
# The keys of a dictionary in Python are generated by a hashing function. 
# The elements of a dictionary are not ordered and they can be changed.
```

```python
new_dict = dict(a='Brijendra',b='Python',c='Machine Learning')
```

```python
new_dict
```

```python
new_dict.keys()
```

```python
new_dict.values()
```

```python
new_dict['a']
```

```python
new_dict['c']
```

```python
for x,y in new_dict.items():
    print(x,y)
```

```python
new_dict.items()
```

## Conversion of dictionary to dataframe

```python
a
```
